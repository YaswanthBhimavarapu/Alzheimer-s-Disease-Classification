```python
import pandas as pd   # for working with tables (DataFrames)
import numpy as np    # for numerical operations
import matplotlib.pyplot as plt  # for basic plots
import seaborn as sns
df = pd.read_csv("../data/alzheimers_disease_data.csv")  #Load the Data
df.head()
```

```python
df.isnull().sum()  #Missing Values
```

```python
df["Diagnosis"].value_counts()  #Target Variable (Diagnosis)
```

```python
plt.hist(df["Age"], bins=20)
plt.xlabel("Age")               #Histogram: Age
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(x="Diagnosis", data=df)    #Count plot: Diagnosis classes
plt.title("Diagnosis Class Distribution")
plt.show()
```

```python
sns.boxplot(x="Diagnosis", y="Age", data=df)
plt.title("Age vs Diagnosis")    #Bivariate: Feature vs Diagnosis
plt.show()
```

```python
sns.boxplot(x="Diagnosis", y="MMSE", data=df)
plt.title("MMSE vs Diagnosis")    #MMSE vs Diagnosis
plt.show()
```

```python
# Drop non-numeric / irrelevant columns
df_corr = df.drop(columns=["DoctorInCharge"])
```

```python
corr = df_corr.corr()
corr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,12))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```
